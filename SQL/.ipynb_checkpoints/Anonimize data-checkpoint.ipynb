```python
%%time

def dupl(df,field,slice_fields=None):
    ids = df[field]
    res = df[ids.isin(ids[ids.duplicated()])].sort_values(field)
    
    if slice_fields!=None:
        res = res[slice_fields].values.tolist()
    
    return res

dupl(  data
     ,'deal_id_cast_as_string'
     ,'deal_id_cast_as_string'
     )
```
